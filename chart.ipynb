```python
from statistics import median


df = pd.read_excel('sales_data.xlsx')
test = df["_SalesTeamID"].value_counts()
median = test.median()

df["_SalesTeamID"].value_counts().plot(kind='bar')
plt.axhline(y=median, color='g')
plt.show()




```
